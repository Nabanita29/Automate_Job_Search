```python
import smtplib
import pandas as pd
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
```

Import the Dataset with the HR contacts and keep the important information

```python
# Import the dataset
df = pd.read_excel('/content/HR Contacts.xlsx')
# Extract relevant columns: Name, Email, Title, and Company
df = df[['Name', 'Email', 'Title', 'Company']]
df.head()
```

 Define the cold email template with placeholders

```python
# Define a professional email template with placeholders
email_template = """
Dear {Name},

I hope you’re doing well. I’m reaching out to express my interest in contributing to {Company} as a 6 months Winter Intern / Software Developer if required to the organization. With a background in Java, Python, JavaScript and NodeJs I am eager to bring value to your team and contribute to {Company}.

Some of my accomplishments include:

• In my previous internship at TakeUForward I designed and developed a web portal end to end for automating the task of note uploading, implemented secure user authentication and authorization solutions using JavaScript and Node.js, and developed RESTful APIs to help conduct stress testing.
• I have authored a research paper ' A GNN-based Novel Approach to Detect Malicious Traffic in Intrusion Detection System' where I deep dived into advanced data processing and machine learning algorithms.

Please find my resume attached for further details. I’d be thrilled to connect and discuss how my skills can benefit {Company}.

Thank you for your time and consideration.

Best regards,
Nabanita Chatterjee
7447471234
"""

# Function to personalize the email for each contact
def create_personalized_email(row):
    return email_template.format(
        Name=row['Name'],
        Title=row['Title'],
        Company=row['Company']
    )

# Test by creating a personalized email for the first contact
first_contact_email = create_personalized_email(df.iloc[0])
print(first_contact_email)
```

```python
import time  # Optional, for adding delay between emails

# Email and SMTP configurations
SMTP_SERVER = 'smtp.gmail.com'
SMTP_PORT = 587
SENDER_EMAIL = 'nabanitachatterjee291002@gmail.com'
SENDER_PASSWORD = 'pmfk deln fpsz brzl'  # Consider using environment variables for security
RESUME_PATH = '/content/nabanita_chatterjee_resume.pdf'  # Update with the actual path to your resume file

# Function to send email with attachment
def send_email(row):
    receiver_email = row['Email']
    receiver_name = row['Name']
    title = row['Title']
    company = row['Company']

    # Create the email
    msg = MIMEMultipart()
    msg['From'] = f"Nabanita Chatterjee <{SENDER_EMAIL}>"
    msg['To'] = receiver_email
    msg['Subject'] = f"Application for Winter Intern / Sofware Developer Position at {company}"

    # Personalize the email body
    body = create_personalized_email(row)
    msg.attach(MIMEText(body, 'plain'))

    # Attach resume
    with open(RESUME_PATH, 'rb') as attachment:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload(attachment.read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', 'attachment', filename="Nabanita_Chatterjee_Resume.pdf")
        msg.attach(part)

    # Send the email
    try:
        server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
        server.starttls()  # Upgrade the connection to a secure encrypted SSL/TLS
        server.login(SENDER_EMAIL, SENDER_PASSWORD)
        server.sendmail(SENDER_EMAIL, receiver_email, msg.as_string())
        server.quit()
        print(f"Email sent successfully to {receiver_name} at {receiver_email}")
    except Exception as e:
        print(f"Failed to send email to {receiver_name}. Error: {e}")

# Send email only to the first contact in the DataFrame
send_email(df.iloc[0])  # This sends the email to the first contact only
```
